```python
from ggplot import *
import pandas as pd
%matplotlib inline
```

```python
surveys_df = pd.read_csv('surveys.csv')

#Ensure all na's are removed before plotting
surveys_df = surveys_df.dropna(axis=0)
```

```python
ggplot(aes(x = 'hindfoot_length', y='weight', color='species_id'), data=surveys_df) + geom_point(alpha=0.2)
```

```python
yearly_species_counts = surveys_df.groupby(['species_id','year','sex'])['record_id'].count().reset_index()
yearly_species_counts['yearly_count'] = yearly_species_counts.record_id

ggplot(aes(x='year', y='yearly_count', color='sex'), data = yearly_species_counts) + geom_line() + \
facet_wrap('species_id', scales='free_y')
```
